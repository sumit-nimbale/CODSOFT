```python
# 03_feature_engineering.ipynb

import pandas as pd

# Load cleaned data
df = pd.read_csv(r"D:\in\Task_2\Dataset\fraud_train.csv")

# Convert to datetime
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['dob'] = pd.to_datetime(df['dob'])

# Time features
df['transaction_hour'] = df['trans_date_trans_time'].dt.hour
df['transaction_day'] = df['trans_date_trans_time'].dt.day
df['transaction_month'] = df['trans_date_trans_time'].dt.month

# Age feature
df['customer_age'] = (
    (df['trans_date_trans_time'] - df['dob']).dt.days // 365
)

# Drop raw datetime columns
df.drop(['trans_date_trans_time', 'dob'], axis=1, inplace=True)

# Save engineered dataset
df.to_csv("/feature_engineered_data.csv", index=False)

print("âœ” feature_engineered_data.csv saved")
```
