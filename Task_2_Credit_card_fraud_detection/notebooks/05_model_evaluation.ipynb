```python
# 05_model_evaluation.ipynb

import pandas as pd
import joblib
from sklearn.metrics import (
    precision_score, recall_score, f1_score,
    roc_auc_score, classification_report
)

# Load data & models
X_test = pd.read_csv(r"D:\in\Task_2\Dataset\fraud_train.csv")
y_test = pd.read_csv(r"D:\in\Task_2\Dataset\fraud_test.csv").squeeze()

lr = joblib.load(r"D:\in\Task_2\mainfiles\logistic_model.pkl")
rf = joblib.load(r"D:\in\Task_2\mainfiles\random_forest_model.pkl")

# Probabilities
lr_probs = lr.predict_proba(X_test)[:, 1]
rf_probs = rf.predict_proba(X_test)[:, 1]

# Default metrics
print("Logistic Regression:")
print("ROC-AUC:", roc_auc_score(y_test, lr_probs))

print("\nRandom Forest:")
print("ROC-AUC:", roc_auc_score(y_test, rf_probs))
```
