```python
# 06_inference_demo.ipynb

import pandas as pd
import joblib

# Load model
model = joblib.load("../models/logistic_model.pkl")

# Example transaction (sample row)
sample = pd.read_csv("../data/X_test.csv").iloc[[0]]

# Predict probability
fraud_prob = model.predict_proba(sample)[0][1]

threshold = 0.30
prediction = "Fraud" if fraud_prob >= threshold else "Legitimate"

print(f"Fraud Probability: {fraud_prob:.4f}")
print(f"Prediction: {prediction}")
```
